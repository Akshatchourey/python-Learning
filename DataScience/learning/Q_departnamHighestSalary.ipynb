```python
import pandas as pd
import numpy as np
```

```python
df1 = {'id':[1,2,3,4,5],
      'name':['jomn','jim','henry','sam','max'],
      'salary':[70000,90000,80000,60000,90000],
      'departmentId':[1,1,2,2,1]}
df1 = pd.DataFrame(df1)
```

```python
df2 = pd.DataFrame({'id':[1,2],
                   'name':['IT', 'Sales']})
df2
```

```python
df2.rename(columns={'id':'departmentId', 'name':'dname'}, inplace=True)
```

```python
df2.head()
```

```python
df3 = pd.merge(df1, df2, on='departmentId')
df3
```

```python
groupdf = df3.groupby('dname')['salary'].max()
groupdf
```

```python
groupdf_salary = df3.groupby('dname')['salary'].transform('max')
groupdf_attempt3 = df3[df3['salary'] == groupdf_salary]
groupdf_attempt3[['dname', 'name', 'salary']]
```
